```python
#performing one hot encoding for top 10 most frequent category
data.X1.value_counts().sort_values(ascending=False).head(20)
```

```python
#similarly for other columns like X1 X3 X4... can be performed taking for X2
#performing one hot encoding for top 10 most frequent category
data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
top_10 =[x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top_10
```

```python
for label in top_10:
    data[label]=np.where(data['X2']==label,1,0)

data[['X2']+top_10].head(40)    
```

```python
#for all categorical variables
def one_hot_top(df,variable,top_x_labels):
    for label in top_x_labels:
        df[variable +'_'+label]=np.where(data[variable]==label,1,0)
#read data again        
data = pd.read_csv('C:\\Users\\Dell\\Python practise\\one hot encoding\\test.csv',usecols=['X1','X2','X3','X4','X5','X6'])

#top 10 frequent for x1 
top_10=[x for x in data.X1.value_counts().sort_values(ascending=False).head(10).index]

#encode X1 into 10 most frequent categories
one_hot_top(data,'X1',top_10)
data.head()
```

```python
#top 10 frequent for x2
top_10=[x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]

#encode X2 into 10 most frequent categories
one_hot_top(data,'X2',top_10)
data.head()
```
