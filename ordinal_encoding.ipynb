we basically handle ordinal category variable by ranking them accordingly


```python
import pandas as pd
import datetime
```

```python
df_today=datetime.datetime.today()
df_list =[df_today - datetime.timedelta(days=x)for x in range(0,10)]
df=pd.DataFrame(df_list)
df.columns=['Day']
df
```

```python
df['Day_of_week']=df['Day'].dt.day_name()
df.head(10)
```

```python
weekday_map={'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,'Friday':5,'Saturday':6,'Sunday':7}
df['Rank']=df.Day_of_week.map(weekday_map)
df.head(10)
```
